```python
import pandas as pd
rows = ['SNR_in', 'PSNR_in', 'SSIM_in', 'PSNR', 'SSIM']
cols = ['F100', 'F200', 'F300', 'F400', 'L100', 'L400']
all_rows = [f"{i}_{x}" for i in range(1,11) for x in rows]
res_df = pd.DataFrame(index=cols, columns=all_rows)

# LOAD DATAFRAME FOR FOCUS MODELS:

for i in range(100,500,100):
  res = []
  for j in range(401,411):

  # Read input data from the first text file
    with open(f"/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_{i}_FOCUS/results/Bmode_training{j}.png/results_input.txt", 'r') as file:
        input_data = file.read().splitlines()
        res += input_data


    # Read output data from the second text file
    with open(f"/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_{i}_FOCUS/results/Bmode_training{j}.png/results.txt", 'r') as file:
        output_data = file.read().splitlines()
        res += output_data

  res_df.loc[f"F{i}"] = res

# LOAD DATAFRAME FOR LINEAR MODELS:

for i in [100,400]:
  res = []
  for j in range(401,411):

  # Read input data from the first text file
    with open(f"/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_{i}_LINEAR/results/Bmode_training{j}.png/results_input.txt", 'r') as file:
        input_data = file.read().splitlines()
        res += input_data


    # Read output data from the second text file
    with open(f"/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_{i}_LINEAR/results/Bmode_training{j}.png/results.txt", 'r') as file:
        output_data = file.read().splitlines()
        res += output_data

  res_df.loc[f"L{i}"] = res
```

```python
final_res = res_df.T
print('--- ALL RESULTS ---')
final_res
```

```python
res_row = ['PSNR', 'SSIM']
all_res_row = [f"{i}_{x}" for i in range(1,11) for x in res_row ]
```

```python
print('--- FINAL OUTPUT RESULTS ---')
final_res.loc[all_res_row]
```
