```python
f100 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_100_FOCUS/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
f200 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_200_FOCUS/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
f300 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_300_FOCUS/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
f400 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_400_FOCUS/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
l100 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_100_LINEAR/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
l200 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_200_LINEAR/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
l300 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_300_LINEAR/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
l400 = pd.read_csv('/content/my_drive/MyDrive/CNRS Research/Deep_DeQuIP_400_LINEAR/log.csv', usecols=['epoch', 'mean_squared_error'], index_col = 0)
```

```python
# selecting only the first 5 epochs for loss comparison, as the loss oscillates back and forth, post 5th epoch.
f100 = f100.iloc[:6]
f200 = f200.iloc[:6]
f300 = f300.iloc[:6]
f400 = f400.iloc[:6]
l100 = l100.iloc[:6]
l200 = l200.iloc[:6]
l300 = l300.iloc[:6]
l400 = l400.iloc[:6]
```

```python
f100 # dataframe for DIVA Model trained on 100 FOCUS IMAGES
```

```python
import matplotlib.pyplot as plt
```

```python
linear = {
    'Deep_DeQuIP_100_LINEAR': l100,
    'Deep_DeQuIP_200_LINEAR': l200,
    'Deep_DeQuIP_300_LINEAR': l300,
    'Deep_DeQuIP_400_LINEAR': l400,
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each dataset
for label, df in linear.items():
    plt.plot(df.index, df['mean_squared_error'], label=label)


plt.title("Comparison of Mean Squared Error over Epochs (LINEAR IMAGES)")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```

```python
focus = {
    'Deep_DeQuIP_100_FOCUS': f100,
    'Deep_DeQuIP_200_FOCUS': f200,
    'Deep_DeQuIP_300_FOCUS': f300,
    'Deep_DeQuIP_400_FOCUS': f400
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each dataset
for label, df in focus.items():
    plt.plot(df.index, df['mean_squared_error'], label=label)


plt.title("Comparison of Mean Squared Error over Epochs (FOCUS IMAGES)")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```

```python
datasets = {
    'Deep_DeQuIP_100_FOCUS': f100,
    'Deep_DeQuIP_200_FOCUS': f200,
    'Deep_DeQuIP_300_FOCUS': f300,
    'Deep_DeQuIP_400_FOCUS': f400,
    'Deep_DeQuIP_100_LINEAR': l100,
    'Deep_DeQuIP_200_LINEAR': l200,
    'Deep_DeQuIP_300_LINEAR': l300,
    'Deep_DeQuIP_400_LINEAR': l400,
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each dataset
for label, df in datasets.items():
    plt.plot(df.index, df['mean_squared_error'], label=label)


plt.title("Comparison of Mean Squared Error over Epochs (COMBINED PLOT)")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```

```python
# Plotting F_100 and L_100 for 25 epochs to show the overshooting of loss. Moreover, losses for LINEAR Images seem to be more noisy, compared to FOCUS Images.
plt.plot(f100, label ='100_FOCUS')
plt.plot(l100, label = '100_LINEAR')
plt.title('Comparison of Loss Convergence of Linear vs Focus')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
