```python
print(f100)
print(f200)
```

```python
import matplotlib.pyplot as plt
```

```python
linear = {
    'Deep_DeQuIP_100_LINEAR': l100,
    #'Deep_DeQuIP_200_LINEAR': l200,
    #'Deep_DeQuIP_300_LINEAR': l300,
    'Deep_DeQuIP_400_LINEAR': l400,
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each dataset
for label, df in linear.items():
    plt.plot(df['epoch'], df['mean_squared_error'], label=label)

plt.yscale('log')
plt.title("Comparison of Mean Squared Error over Epochs (LINEAR IMAGES)")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```

```python
focus = {
    'Deep_DeQuIP_100_FOCUS': f100,
    'Deep_DeQuIP_200_FOCUS': f200,
    'Deep_DeQuIP_300_FOCUS': f300,
    'Deep_DeQuIP_400_FOCUS': f400
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each dataset
for label, df in focus.items():
    plt.plot(df['epoch'], df['mean_squared_error'], label=label)

plt.yscale('log')
plt.title("Comparison of Mean Squared Error over Epochs (FOCUS IMAGES)")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```

```python
# Plotting F_100 and L_100 for 25 epochs to show the overshooting of loss. Moreover, losses for LINEAR Images seem to be more noisy, compared to FOCUS Images.
plt.plot(f100['epoch'], f100['mean_squared_error'], label ='100_FOCUS')
plt.plot(l100['epoch'],l100['mean_squared_error'], label = '100_LINEAR')
plt.title('Comparison of Loss Convergence of Linear vs Focus')
plt.yscale('log')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

```python
# F_400 vs L_400 Images
plt.plot(f400['epoch'], f400['mean_squared_error'], label ='400_FOCUS')
plt.plot(l400['epoch'],l400['mean_squared_error'], label = '400_LINEAR')
plt.title('400_Linear vs 400_Focus')
plt.yscale('log')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
